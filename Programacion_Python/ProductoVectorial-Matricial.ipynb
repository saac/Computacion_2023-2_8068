```python
def ProdInt(v1,v2):
    a = 0.0
    if len(v1) != len(v2):
        print("Vectores de Diferente dimension")
#         print len(v1), len(v2)
    else:
        dim = len(v1)
        
        for i in range(dim):
            a = a + v1[i]*v2[i]
            
    return a
```

```python
def renglones(X):
    rens = []
    for i in range(len(X)):
        rens.append(X[i])
    
    return rens
```

```python
def columnas(X):
    cols = []
    for i in range(len(X[0])):
        vT = []
        for j in range(len(X)):
            vT.append(X[j][i])
        cols.append(vT)
                   
    return cols
```
